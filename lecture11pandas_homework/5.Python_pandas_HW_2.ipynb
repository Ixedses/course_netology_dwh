### Задание 1

```python
import pandas as pd

movies=pd.read_csv('11pandas/ml-latest-small/movies.csv')
ratings=pd.read_csv('11pandas/ml-latest-small/ratings.csv')

joined = ratings.merge(movies, on='movieId', how='left')

def rating_classification(row):
    if  row['rating'] <= 2.0:
        return 'низкий рейтинг'
    elif row['rating'] <= 4.0 and row['rating'] > 2.0:
        return 'средний рейтинг'
    elif row['rating'] >= 4.5:
        return 'высокий рейтинг'
    
joined['class'] = joined.apply(rating_classification, axis=1)
joined[['title','rating','class']].head()
```

### Задание 2

```python
import pandas as pd
keywords=pd.read_csv('11pandas/ml-latest-small/keywords.csv')


geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def city_geolocation(row):
    if ('москва'in row['keyword']) or ('тула' in row['keyword']) or ('ярославль'in row['keyword']):
        return 'Центр'
    elif ('петербург' in row['keyword']) or ('псков' in row['keyword']) or ('мурманск' in row['keyword']):
        return 'Северо-Запад'
    elif ('владивосток' in row['keyword']) or ('сахалин' in row['keyword']) or ('хабаровск' in row['keyword']):
        return 'Дальний Восток'
    else:
        return 'undefined'


    
keywords['region'] = keywords.apply(city_geolocation, axis=1)
keywords[(keywords['region']!='undefined')]
```
