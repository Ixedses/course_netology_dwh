### Задание 3

```python
import pandas as pd

rating = pd.read_csv('12pandas/ml-latest-small/ratings.csv')

rating_count = rating.groupby('userId').rating.agg(['count']).reset_index()
count_more_100 = rating_count[rating_count['count'] > 100]
timestamp = rating.groupby('userId').timestamp.agg(['max','min']).reset_index()
timestamp_avg = count_more_100.merge(timestamp, how ='left')

def average(row):
    timestamp_mean=(row['max']-row['min'])
    return timestamp_mean

timestamp_avg['avg'] = timestamp_avg.apply(average, axis=1)

timestamp_avg.head(300)
```
