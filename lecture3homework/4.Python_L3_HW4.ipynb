```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
#Задание 1
# Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру
def number_to_name():
    doc_number = input ('Введите номер документа: ')
    for document in documents:
        if doc_number == document['number']:
            a=document['name']
            return (f'Владелец документа: {a}')
    return ('Документ не найден в базе')

# Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
def documents_to_directories():
    doc_number = input('Введите номер документа: ')
    for key in directories:
        if doc_number in directories[key]:
            return (f'Документ хранится на полке: {key}')
    return ('Документ не найден в базе')

# Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам
def info_total():
    for key,value in directories.items():
        for document in documents:
            doc_type = document['type']
            doc_number = document['number']
            name = document['name']
            if  doc_number in value :                
                print (f'№: {doc_number}, тип: {doc_type}, владелец: {name}, полка хранения: {key}')
    
                
# Пункт 4. Пользователь по команде “as” может добавить новую полку
def add_shelf():
    shelf_number = input('Введите номер полки: ')
    directories_copy = directories.copy()
    for key in directories_copy:
        if  shelf_number  in directories.keys():
            keys =(','.join(list (directories_copy.keys())))
            return(f'Такая полка уже существует. Текущий перечень полок: {keys}')
        else :
            keys =(','.join(list (directories_copy.keys())))
            directories.update({shelf_number:[]})
            return(f'Полка добавлена. Текущий перечень полок: {keys},{shelf_number}')
        
# Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)        
def del_shelf():
    number = input('Введите номер полки: ')
    if number in directories:
        if len(directories[number]) > 0:
            print( 'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + ', '.join(directories.keys()))
        elif len(directories[number]) <= 0:
            del directories[number]
            print( 'Полка удалена. Текущий перечень полок: ' + ', '.join(directories.keys()))
    else:
        print ('Такой полки не существует. Текущий перечень полок:' + ', '.join(directories.keys()))
        
        
        

#Задание 2 (необязательное)
#Пункт 1. Пользователь по команде “ad” может добавить новый документ в данные
def add_documents():
    doc_number = input ('Введите номер документа: ')
    doc_type = input ('Введите тип документа: ')
    name = input ('Введите владельца документа: ')
    shelf_number   = input ('Введите полку для хранения: ')
    if  shelf_number not in directories.keys():
        print ('Такой полки не существует. Добавьте полку командой as.', '\n'
              'Текущий список документов:')
        info_total()
    else:
        directories[shelf_number].append(doc_number)
        documents.append({'type': doc_type, 'number': doc_number, 'name': name})
        print ('Документ добавлен. Текущий список документов:')
        info_total()
            
#Пункт 2. Пользователь по команде “d” может удалить документ из данных   
#Пункт 3. Пользователь по команде “m” может переместить документ с полки на полку
```

```python
def main ():    
    while True:
        command = input("Введите команду: ")
        if command == 'p':
            print(number_to_name())
        elif command == 's':
            print(documents_to_directories())
        elif command == 'l': 
            info_total()
        elif command == 'as':
            print(add_shelf())
        elif command == 'ds':
            del_shelf()
        elif command == "ad":
            add_documents()
        elif command == "q":        
            print('Работа окончена.')
            break
        else:
            print ('Введена некорректная команда')
            
            
main()
```
